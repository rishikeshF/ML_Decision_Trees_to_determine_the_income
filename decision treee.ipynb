here, we are looking for records having '?' as its values so that we can get rid of them

```python
data = data[data['native.country'] != '?']
data = data[data['occupation'] != '?']
```

```python
# Data cleaning is done here, now we will proceed to data preparation. We need to convert the categorical values in
# the form understandable to the machine
```

```python
from sklearn import preprocessing
```

```python
data_categorical = data.select_dtypes(include=['object'])
```

```python
data_categorical.head()
```

These are the categorical values in our database. Now, we need to remove these values from here and replace them with numerical
values equivalent to these categorical values.

So, firstly, we will find numerical equivalent for these values

```python
le= preprocessing.LabelEncoder()
data_categorical = data_categorical.apply(le.fit_transform)
```

```python
data_categorical.head()
```

so this is how we have converted the categorical values into numerical equivalent. now, we need to delete the 
categorical columns and insead concat these new numerical equivalents of those categorical values

```python
data = data.drop(data_categorical.columns, axis = 1)
data = pd.concat([data, data_categorical], axis = 1)
```

```python
data.info()
```

```python
data.income = data['income'].astype('category')
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = data.drop('income', axis = 1)
```

here we are inserting the dataset in variable x excluding the column income which is going to be our y variable.

```python
y =data['income']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = .3)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
data_default = DecisionTreeClassifier(max_depth = 5)
```

```python
data_default.fit(x_train, y_train) 
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
y_pred_default = data_default.predict(x_test)
```

```python
print(classification_report(y_test,y_pred_default))
```

```python
print(confusion_matrix(y_test,y_pred_default))
```

```python
print(accuracy_score(y_test,y_pred_default))
```

tuning with max_depth

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import GridSearchCV
```

```python
n_folds= 5
```

```python
parameters = {'max_depth': range(1,40)}
```

```python
dtree = DecisionTreeClassifier(criterion = 'gini')
tree = GridSearchCV(dtree, parameters, cv = n_folds, scoring="accuracy")
tree.fit(x_train,y_train)
```

```python
scores = tree.cv_results_
```

```python
pd.DataFrame(scores).head()
```

```python
plt.figure()
plt.plot(scores["param_max_depth"],
        scores["mean_train_score"],
        label="training_accuracy")
plt.plot(scores["param_max_depth"],
        scores["mean_test_score"],
        label = "test_accuracy")
plt.xlabel("max_depth")
plt.ylabel("Accuracy")
plt.legend()
plt.show()
```

Here, the model is over-fitting wiht increasing depth
